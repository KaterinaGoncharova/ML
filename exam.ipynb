```python
artists=['ABBA', 'The Beatles', 'Backstreet Boys', 'Elvis Costello']
text = pd.read_csv('songdata.csv')
text = text[text['artist'].isin(artists) == True]
text['text'] = text['text'].apply(lambda line: line.replace('\n', ''))
```

```python
text.head(200)
```

Ищу минимальную длину песни, чтобы всех поравнять.

```python
min_count = min(len(text[text['artist'] == 'ABBA']), len(text[text['artist'] == 'The Beatles']),  len(text[text['artist'] == 'Backstreet Boys']), len(text[text['artist'] == 'Elvis Costello']))
print(min_count)
```

Создаю датафрейм с одинаковыми текстами песен для исполнителей.

```python
Abba = text[text['artist'] == 'ABBA'][:min_count]
Beatles = text[text['artist'] == 'The Beatles'][:min_count]
Boys = text[text['artist'] == 'Backstreet Boys'][:min_count]
Elvis = text[text['artist'] == 'Elvis Costello'][:min_count]
text = pd.concat([Abba, Beatles, Boys, Elvis])
text.head(200)
```

Делю на train и test

```python
X_train, X_test, y_train, y_test = train_test_split(text['text'] , text['artist'], test_size=0.2, random_state=40)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
#stops = open('stops.txt', 'r', encoding='utf-8')
#stops.read()
```

```python
vec = CountVectorizer()
X_train_mat = vec.fit_transform(X_train)
X_test_mat = vec.transform(X_test) #матрица с кол-вом токенов
```

использую DummyClassifier ,который угадывает исполнителя. (в качестве baseline)

```python
DC = DummyClassifier()
DC.fit(X_train_mat, y_train)
y_pred_mat = DC.predict(X_test_mat)
print(classification_report(y_test, y_pred_mat, target_names=['ABBA', 'The Beatles', 'Backstreet Boys', 'Elvis Costello']))
```

precision - (Точность показывает, сколько из предсказанных позитивных объектов, оказались действительно позитивными.)

recall -(Полнота показывает, сколько от общего числа реальных позитивных объектов, было предсказано, как позитивный класс.)


Если векторизовать не слова, а n-граммы, которые будут длиной от 2 до 3

```python
v = CountVectorizer(ngram_range=(2,3))
X_train2 = v.fit_transform(X_train)
X_test2 = v.transform(X_test)
```

```python
DC = DummyClassifier()
DC.fit(X_train2, y_train)
y_pred2 = DC.predict(X_test2)
print(classification_report(y_test, y_pred2, target_names=['ABBA', 'The Beatles', 'Backstreet Boys', 'Elvis Costello']))
```
